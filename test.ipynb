```python
from gqlalchemy import Memgraph

# Make a connection to the database
memgraph = Memgraph(host='127.0.0.1', port=7687)

# Delete all nodes and relationships
query = "MATCH (n) DETACH DELETE n"

# Execute the query
memgraph.execute(query)

# Create a node with the label FirstNode and message property with the value "Hello, World!"
query = """CREATE (n:FirstNode)
           SET n.message = '{message}'
           RETURN 'Node '  + id(n) + ': ' + n.message AS result""".format(message="Hello, World!")

# Execute the query
results = memgraph.execute_and_fetch(query)

# Print the first member
print(list(results)[0]['result'])
```

```python
create_query = """
CREATE (a:Person {name: 'Asad', role: 'Jr AI ML Developer'})
CREATE (b:Person {name: 'Manager', role: 'Tech Lead'})
CREATE (s1:Skill {name: 'LLMs'})
CREATE (s2:Skill {name: 'RAG'})
CREATE (s3:Skill {name: 'Graphs'})
CREATE (a)-[:KNOWS]->(s1)
CREATE (a)-[:LEARNING]->(s2)
CREATE (a)-[:LEARNING]->(s3)
CREATE (b)-[:KNOWS]->(s2)
CREATE (b)-[:KNOWS]->(s3)
CREATE (a)-[:REPORTS_TO]->(b);
"""

memgraph.execute(create_query)
print("Sample graph created.")
```
